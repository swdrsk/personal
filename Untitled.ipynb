```python
import networkx as nx
import numpy as np

#ランダムグラフを生成（WSモデル）
G = nx.watts_strogatz_graph(n=200,k=5,p=0.1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#グラフを可視化
nx.draw(G)
plt.show()
```

```python
#次数を表示（返り値は辞書型）
print nx.degree(G) # or G.degree()
```

```python
print nx.average_shortest_path_length(G)
```

```python
print nx.average_degree_connectivity(G)
```

```python
print nx.average_neighbor_degree(G)
```

```python
print nx.average_node_connectivity(G)
```

```python
print np.average(nx.degree(G).values())
```
